TensorFlow-Lite, TensorFlow modellerinin mobil cihazlarda, gömülü sistemlerde ve IoT cihazlarda daha verimli çalıştırılması için özelleştirilmiş bir araç setidir. İki ana birimden oluşur:


1.   **TensorFlow-Lite Yorumlayıcı:** Optimize edilmiş modelleri farklı donanımlar üzerinde çalıştırır. Sade ve hızlı olarak tasarlanmıştır. TensorFlow-Lite modeli üretmek için tf-lite yorumlayıcısı kullanılır.  Yorumlayıcı minimum yük, başlatma ve uygulama gecikmelerini kesinleştirmek için statik bir grafik ve özel bir bellek ayırıcı kullanır. Bununla birlikte tf-lite yorumlayıcısı şimdilik sınırlı sayıda tf operatörünü desteklemektedir. Bu bazı modellerin tf-lite ile çalışması için ek işlemler gerektirebileceği anlamına gelmektedir.

2.   **TensorFlow-Lite Dönüştürücü:** TensorFlow modellerinin tf-lite tarafından kullanılması için dönüştürülmesi gerekebilir. Böylece binary dosya boyutu ve performans optmizasyonları yapılabilir. 
TensorFlow-Lite dönüştürücü, tf modellerini optimize edilmiş FlatBuffer formatına dönüştürmek için kullanılır. Böylece oluşturulan “.tflite” uzantılı modeller tf-lite tarafından kullanılabilir. Dönüştürücünün desteklediği formatlar; SavedModels, Frozen Graph Def, tf.lite, tf.Session ile alınan herhangi bir model (sadece Python-API ile kullanılabilir.)

## **Temel Özellikler**

1.   Cihaz üzerinde makine öğrenmesi çalıştırmak için cihaz ayarları, cihaz üzerindeki uygulamalar için optimize edilmiş ve küçük binary boyutta bir dizi kernel işlemini destekler.

3.   Birçok programlama dili için API sunar.

## **Desteklenen Platformlar**

1. Android
Java ya da C++ API  Kullanılarak tf-lite yorumlayıcısı uygulanabilir. Java API doğrudan Android Activity bölümlerinde kullanılabilir. C++ API daha fazla esneklik ve hız sağlar, ancak Java ile C++ arasında veri aktarımı için JNI Wrapper kodlanması gerekebilir.

2. iOS
TensorFlow-Lite Swift ve Objective-C ile yazılan native iOS kütüphanelerinin kullanımını destekler.

3. Linux
C++ ve Python API kullanılarak linux platformlarında çalıştırılabilir.

**Not:** TensorFlow-Lite geliştirme sürecinde olduğundan dolayı gelecekte mevcut operatörlerde ve API’lerde değişiklik/güncelleme yapılabilir.

## **Veri Setlerinin Alınması**

Veri setlerinin obje tespit etme algoritması eğitimine başlamadan önce veri setlerindeki her bir resmin makine tarafından öğrenmesini istediğiniz yerleri belirtmeniz gerekmektedir. Bu kısacası 2 boyutlu görsellerde bir noktayı (x,y) şeklinde belirtmekteyiz. Makine bunu verilen iki noktanın bir başlangıç noktasını bir de bitiş yoktasını tutmaktadır. Bu noktalarla birlikte etiketleri varsa eğer onları da alması gerekmektedir. 

Tüm bunlar iki tip dosyada depolanabilmektedir:

1. YOLO formatında yani bir 'txt' dosyası içerisinde boşluklarla ayrılmış etiket, başlangıçta x ve y noktalarından sonra bitişte x ve y noktalarının normalize edilmiş hallerini depolamaktadır. Bu yöntem algoritma için anlaşılması basittir.

2. Pascal_Voc formatında ise bir 'xml' dosyası içerisinde yer alan elementler içinde görüntünün etiketi ve o etiketin ismi gibi verilerle birlikte alınması gereken x ve y noktaları görüntüye oranla olduğu gibi depolanmaktadır. Yani normalize edilmeden depolanmaktadır.

### **Örnekler**

```
YOLO Formatı için:
# 1 0.5841911764705883 0.535625 0.030147058823529412 0.04375
```



```
Pascal Voc Formatı için:

	<folder>images</folder>

	</source>
	<size>
		<width>1360</width>
		<height>800</height>
		<depth>3</depth>
	</size>
	<segmented>0</segmented>

		<name>danger</name>
		<pose>Unspecified</pose>
		<truncated>0</truncated>
		<difficult>0</difficult>
		<bndbox>
			<xmin>774</xmin>
			<ymin>411</ymin>
			<xmax>815</xmax>
			<ymax>446</ymax>
		</bndbox>
	</object>
</annotation>
```

````python
import os

train_path = os.path.join('/content','drive','MyDrive','DATASETS','images','train') 
test_path = os.path.join('/content','drive','MyDrive','DATASETS','images','test') 
train_data = object_detector.DataLoader.from_pascal_voc(train_path, train_path, ['prohibitory', 'danger', 'mandatory', 'other'])
validation_data = object_detector.DataLoader.from_pascal_voc(test_path, test_path, ['prohibitory', 'danger', 'mandatory', 'other'])
print(train_data)
````
